```python

import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0,2,1000)
f =2
sine_wave = np.sin(2*np.pi*f*t)
plt.figure(figsize =(12,4))
plt.plot(t,sine_wave)
plt.title('sinewave')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.show()
```

```python
t =np.linspace(-2,2,1000)
step = np.heaviside(t,0.5)
plt.figure(figsize =(12,4))
plt.plot(t,step,'-g',linewidth =2)
plt.title('unit step signal')
plt.ylim(-0.2,1.2)
plt.show()
```

```python
step = np.heaviside(t, 0.5)
plt.figure(figsize=(12, 4))
plt.plot(t, step, '-g', linewidth=2)
plt.title('Unit Step Signal')
plt.ylim(-0.2, 1.2)
plt.grid(True)
plt.show()
```

```python
exp_signal = np.exp(t) * (t >= 0)
plt.figure(figsize=(12, 4))
plt.plot(t, exp_signal, '-m', linewidth=2)
plt.title('Exponential Signal')
plt.ylim(-0.2, np.max(exp_signal)+0.5)
plt.grid(True)
plt.show()
```

```python
ramp = t * (t >= 0)
plt.figure(figsize=(12, 4))
plt.plot(t, ramp, '-b', linewidth=2)
plt.title('Ramp Signal')
plt.ylim(-0.2, np.max(ramp)+0.5)
plt.grid(True)
plt.show()
```

```python
t = np.linspace(-2, 2, 1000)
impulse = np.zeros_like(t)
impulse[np.abs(t) < 0.001] = 1
plt.figure(figsize=(12, 4))
plt.stem(t, impulse, linefmt='r', markerfmt='ro', basefmt=" ")
plt.title('Unit Impulse Signal (Approximation)')
plt.ylim(-0.2, 1.2)
plt.grid(True)
plt.show()
```

```python
n = np.arange(-10, 11)
impulse_digital = np.zeros_like(n)
impulse_digital[n == 0] = 1
plt.figure(figsize=(12, 4))
plt.stem(n, impulse_digital, linefmt='r', markerfmt='ro', basefmt=" ")
plt.title('Digital Unit Impulse Signal δ[n]')
plt.ylim(-0.2, 1.2)
plt.grid(True)
plt.show()
```

```python
n = np.arange(-10, 11)
sign_signal = np.sign(n)
plt.figure(figsize=(12, 4))
plt.stem(n, sign_signal, linefmt='b', markerfmt='bo', basefmt=" ")
plt.title('Discrete-Time Signum Signal sgn[n]')
plt.ylim(-1.2, 1.2)
plt.grid(True)
plt.show()
```
