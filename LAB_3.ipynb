```python
#for loop - iterating over a range
print("for loop-Counting 1 to 5:")
for i in range (1,6):
  print(f"Count: {i}")

#for loop- iterating over a list
print("\n for loop-Iterating through fruits:")
fruits=["apple","banana","orange","grape"]
for fruit in fruits:
  print(f"I like {fruit}")
```

```python
print ("\nFor loop - Iterating through characters:")
word = "Python"
for letter in word:
  print (f"Letter:{letter}")

#while loop
print("\n for loop-countdown")
count=5
while count>0:
  print(f"Countdown: {count}")
  count-=1 #decreae count by 1
  print("Blast off")
```

```python
#break and continue
print("\n Break and continue Example:")
for num in range(1,11):
  if num==5:
    continue
  if num==8:
    break
  print(f"Number: {num}")
```

```python
print("\n"+"="*50)
print("lists")
print("="*50)

#creating lists
numbers=[1,2,3,4,5]
mixed=[1,"hello",3.14,True]
colors=["red","green","blue"]

print(f"Numbers: {numbers}")
print(f"Mixed lists: {mixed}")
print(f"Colors: {colors}")

print(f"\nFirst Color: {colors[0]}")
print(f"\n Last color:{colors[1]}")

colors.append("yellow")
print(f"After append: {colors}")

colors.insert(1,"orange")
print(f"After insert: {colors}")

colors.remove("green")
print(f"After remove : {colors}")
#slicing lists
print(f"\n First 2 colors: {colors[:2]}")
print(f"\n First 2 colors: {colors[-2:]}")

#list comprehension
squares=[x**2 for x in range(1,6)]
print(f"\nSquares: {squares}")

#dictionaries
print("\n"+"="*50)
print("DICTIONARIES(key-value pairs)")
print("="*50)
```

```python
#creating a dictionary
person={
    "name":"John",
    "age": 30,
    "City":"New york",
    "is_student": False
}
print(f"person dictionary:{person}")
print(f"Age:{person.get('age')}")
```
