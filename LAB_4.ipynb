```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(-5,5,0.01)
unit_step = (t>=0).astype(float)
unit_impulse=(t==0).astype(float)
ramp = t*(t>=0).astype(float)
sine_wave= np.sin(2*np.pi*1*t)
exponential =np.exp(-0.5*t)*(t>=0).astype(float)
fig,axs =plt.subplots(5,1,figsize =(10,10))
axs[0].plot(t,unit_step,linewidth=2,color='b')
axs[0].set_title('Unit Step Signal')
axs[0].grid(True)

axs[1].plot(t,unit_impulse,linewidth=3,color='g')
axs[1].set_title('Unit Implulse signal')
axs[1].grid(True)

axs[2].plot(t,ramp,linewidth=2,color='y')
axs[2].set_title('Ramp Signal')
axs[2].grid(True)

axs[3].plot(t,sine_wave,linewidth=2,color='r')
axs[3].set_title('Sine wave ')
axs[3].grid(True)

axs[4].plot(t,exponential,linewidth=2,color='b')
axs[4].set_title('Exponential decay')
axs[4].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-5, 5, 0.01)
x1 = np.sin(2 * np.pi * t)
x2 = np.cos(2 * np.pi * t)
y_cont = x1 + x2

n = np.arange(-20, 21)
x1_d = np.sin(0.2 * np.pi * n)
x2_d = np.cos(0.2 * np.pi * n)
y_disc = x1_d + x2_d

plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(t, x1, label='x1(t)')
plt.plot(t, x2, label='x2(t)')
plt.plot(t, y_cont, label='x1(t)+x2(t)')
plt.title('Continuous-Time Signal Addition')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.stem(n, x1_d, label='x1[n]')
plt.stem(n, x2_d, label='x2[n]')
plt.stem(n, y_disc, label='x1[n]+x2[n]')
plt.title('Discrete-Time Signal Addition')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
